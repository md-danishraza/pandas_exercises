### Step 1. Import the necessary libraries

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Create the DataFrame that should look like the one below.

```python


raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
        'female': [0, 1, 1, 0, 1],
        'age': [42, 52, 36, 24, 73],
        'preTestScore': [4, 24, 31, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70]}

data = pd.DataFrame(raw_data)
data
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
sns.set_theme(style='darkgrid')
sns.scatterplot(x='preTestScore',y='postTestScore',data=data,size='age');
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sns.set_theme(style='darkgrid')

sns.scatterplot(x='preTestScore',y='postTestScore',data=data,size=4.5*data['postTestScore'],hue='female')

```

### BONUS: Create your own question and answer it.

```python
plt.figure(figsize=(10,6))
plt.xlabel('Pre Test Score')
plt.ylabel('Post Test Score')
plt.title('Scatter Plot of Pre Test Score vs Post Test Score with Size Determined by Age')
plt.scatter(data['preTestScore'],data['postTestScore'],s=data['age'])
```
